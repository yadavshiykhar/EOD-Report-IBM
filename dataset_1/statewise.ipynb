```python
#Plotting the Bar Graph 
df1 = df.head()
import matplotlib.pyplot as plt
plt.bar(df1.Date,df1.PositiveCases,color=['g'])
plt.xlabel('Date')
plt.ylabel('PostiveCases')
plt.title('Bar Graph')
plt.show()
```
